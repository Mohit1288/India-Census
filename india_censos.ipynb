# AGRICULTURE DEPENDICY OF STATE

```python
df['ag_dependcy']=round((df['Agricultural_Workers']/df['Workers'])*100)
```

```python
Agriculture_dependency=df.groupby("State_name")['ag_dependcy']
Agriculture_dependency.first().sort_values(ascending=False)
```

# To find sex ratio

```python
df['sex_ratio'] = round((df['Female']/df['Male'])*100)
```

 # To find literacy

```python
df['literacy_rate'] = round((df['Literate']/df['Population'])*100)
```

```python
df.drop(columns=['Male','Female','Literate','District_name','District_code'],inplace=True)
```

Show in table

```python


df.groupby("State_name")['literacy_rate','sex_ratio'].sum()
```

# Sort the values accoroding to Literacy_rate

```python
rk=df.groupby("State_name")["literacy_rate"]
rk.first().sort_values(ascending=False)
```

# show the litercy_rate in bar plot

```python
fig = plt.figure(figsize = (30, 10))

# creating the bar plot
plt.bar(df.State_name, df.literacy_rate,)
```

# Sort the values accoroding to sex_ratio

```python

tt=df.groupby("State_name")["sex_ratio"]
tt.first().sort_values(ascending=False)

```

# show the sex ratio in bar plot

```python
 fig = plt.figure(figsize = (30, 5))

# creating the bar plot
plt.bar(df.State_name, df.sex_ratio, color ='maroon')
```
